```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
cosine_sq_no = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/acc/const/run-train_clean_100_pytorch_train_cosine_square_no_decay-tag-main_acc.csv")
cosine_sq_no_baseline = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/acc/const/run-train_clean_100_pytorch_train_cosine_square_no_decay_baseline-tag-main_acc.csv")

cosine_sq_no_auto = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/acc/auto/run-train_clean_100_pytorch_train_cosine_square_no_decay_dynamic-tag-main_acc.csv")
cosine_sq_no_auto_baseline = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/acc/auto/run-train_clean_100_pytorch_train_cosine_square_no_decay_dynamic_baseline-tag-main_acc.csv")

cosine_sq_no_loss = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/loss/const/run-train_clean_100_pytorch_train_cosine_square_no_decay-tag-main_loss_oracle.csv")
cosine_sq_no_loss_baseline = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/loss/const/run-train_clean_100_pytorch_train_cosine_square_no_decay_baseline-tag-main_loss_oracle.csv")

cosine_sq_no_loss_auto = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/loss/auto/run-train_clean_100_pytorch_train_cosine_square_no_decay_dynamic-tag-main_loss_oracle.csv")
cosine_sq_no_loss_auto_baseline = pd.read_csv("/home/neo/MS/ICASSP_2020/FA/results/loss/auto/run-train_clean_100_pytorch_train_cosine_square_no_decay_dynamic_baseline-tag-main_loss_oracle.csv")
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,10))
ax[0][0].plot(cosine_sq_no["Step"].tolist(), cosine_sq_no["Value"].tolist(), color="red", linestyle='-.')
ax[0][0].plot(cosine_sq_no_baseline["Step"].tolist(), cosine_sq_no_baseline["Value"].tolist(), color="blue", linestyle=':' )

ax[0][1].plot(cosine_sq_no_auto["Step"].tolist(), cosine_sq_no_auto["Value"].tolist(), color="red", linestyle='-.')
ax[0][1].plot(cosine_sq_no_auto_baseline["Step"].tolist(), cosine_sq_no_auto_baseline["Value"].tolist(), color="blue", linestyle=':' )

ax[1][0].plot(cosine_sq_no_loss["Step"].tolist(), cosine_sq_no_loss["Value"].tolist(), color="red", linestyle='-.')
ax[1][0].plot(cosine_sq_no_loss_baseline["Step"].tolist(), cosine_sq_no_loss_baseline["Value"].tolist(), color="blue", linestyle=':' )

ax[1][1].plot(cosine_sq_no_loss_auto["Step"].tolist(), cosine_sq_no_loss_auto["Value"].tolist(), color="red", linestyle='-.')
ax[1][1].plot(cosine_sq_no_loss_auto_baseline["Step"].tolist(), cosine_sq_no_loss_auto_baseline["Value"].tolist(), color="blue", linestyle=':' )

ax[0][0].set_xlabel("Iterations", fontsize=40)
ax[0][1].set_xlabel("Iterations", fontsize=40)

ax[0][0].set_ylabel("Accuracy", fontsize=40)
ax[0][1].set_ylabel("Accuracy", fontsize=40)

ax[1][0].set_xlabel("Iterations", fontsize=40)
ax[1][1].set_xlabel("Iterations", fontsize=40)

ax[1][0].set_ylabel("Alignment\nLoss", fontsize=40, labelpad=20)
ax[1][1].set_ylabel("Alignment\nLoss", fontsize=40, labelpad=20)

ax[0][0].tick_params(axis='both', which='major', labelsize=15)
ax[0][1].tick_params(axis='both', which='major', labelsize=15)
ax[1][0].tick_params(axis='both', which='major', labelsize=15)
ax[1][1].tick_params(axis='both', which='major', labelsize=15)

ax[0][0].set_title("(a)", fontsize=40)
ax[0][1].set_title("(b)", fontsize=40)
ax[1][0].set_title("(c)", fontsize=40)
ax[1][1].set_title("(d)", fontsize=40)

cmap = plt.cm.coolwarm
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='red', lw=6, linestyle='-.'),
                Line2D([0], [0], color='blue', lw=6, linestyle=':')]

ax[0][1].legend(custom_lines, ['With Alignment Loss', 'Without Alignment Loss'], fontsize=20, loc="upper center", bbox_to_anchor=(0.89, 1.4))


fig.tight_layout()

plt.savefig("/home/neo/MS/ICASSP_2020/FA/results/acc_loss.png", dpi=200)
```
