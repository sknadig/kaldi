```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
```

```python
a = np.zeros((27))
a[10:17]=1
```

```python
b = gaussian(np.arange(27),13,1)
```

```python
c = np.zeros((27))
c[13] = 1
```

```python
plt.figure(figsize=(40,10))
plt.plot(a, color="blue", linestyle='--', marker='o', markersize=40, label="Square", linewidth=10)
plt.plot(b, color="green", linestyle='-', marker='v', markersize=40,  label="Gaussian", linewidth=10)
plt.plot(c, color="red", linestyle='-.', marker='s',  markersize=40, label="Impulse", linewidth=10)
plt.legend(prop={'size': 80})
plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
labels = [str("") for ele in np.arange(27)]

labels[10] = "k"
labels[16] = "l"
labels[13] = "c"

plt.xticks(np.arange(27), labels=labels, fontsize=80)
plt.tight_layout()
plt.savefig("/home/neo/MS/ICASSP_2020/FA/results/alignment_type.png", dpi=200)
```
